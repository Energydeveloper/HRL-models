```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure that you've loaded the DataFrame 'df' before calling this function
df = pd.read_csv('evaluation_data.csv')  # Make sure this path is correct

def plot_battery_states(df):
    # Prepare time axis based on the number of entries in the DataFrame
    t = np.arange(len(df['Battery States']))
    
    # Create a plot with specific dimensions and style settings
    plt.figure(figsize=(20, 15))
    plt.rc('font', family='Times New Roman', size=55, weight='bold')  # Set font style globally

    # Create the primary axis
    ax1 = plt.gca()
    ax1.plot(t, df['Battery States'], 'r-', label="Energy-Battery", linewidth=2)
    ax1.set_ylabel("Energy-Battery Energy (p.u.)")
    ax1.set_xlabel("Time Step")
    ax1.grid(True, linestyle='--')  # Add a grid with dashed lines

    # Create a secondary axis to plot another dataset
    ax1_r = ax1.twinx()  # This enables having two y-axes
    ax1_r.plot(t, df['Power Battery States'], 'b-', label="Power-Battery", linewidth=2)
    ax1_r.set_ylabel("Power-Battery Energy (p.u.)")

    # Combine legends from both axes
    lns, labs = ax1.get_legend_handles_labels()
    lns_r, labs_r = ax1_r.get_legend_handles_labels()
    ax1.legend(lns + lns_r, labs + labs_r, loc='upper left')

    # Adjust layout to make sure everything fits without overlapping
    plt.tight_layout()

    # Save the plot as a PDF file
    plt.savefig("battery_states.pdf", format="pdf")

    # Show the plot
    plt.show()

# Call the function to plot the data
plot_battery_states(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the DataFrame
df = pd.read_csv('evaluation_data (3).csv')  # Ensure this path is correct

# Generate timestamps assuming each row is 30 minutes apart, starting from midnight
start_time = '2023-01-01 00:00:00'  # Adjust the start date as necessary
df['Timestamp'] = pd.date_range(start=start_time, periods=len(df), freq='30T')

def plot_evaluation(df):
    """
    Plot the evaluation results using the prepared DataFrame.
    """
    t = df['Timestamp']
    battery_states = df['Battery States']
    power_battery_states = df['Power Battery States']
    grid_actions = df['Grid Actions']
    tariffs = df['Tariffs']
    load_demands = df['Load Demands']
    load_deferrals = df['Load Deferrals']

    plt.figure(figsize=(40, 10))
    plt.subplots_adjust(wspace=0.5)  # Increase horizontal spacing between subplots

    plt.rc('font', family='Times New Roman', size=44, weight='bold')

    ax1 = plt.subplot(1, 3, 1)
    ax1.plot(t, battery_states, 'r-', label="Energy-Battery (p.u.)", linewidth=2)
    ax1.set_ylabel("Energy-Battery Power (p.u.)")
    ax1_r = ax1.twinx()
    ax1_r.plot(t, power_battery_states, 'b-', label="Power-Battery (p.u.)", linewidth=2)
    ax1_r.set_ylabel("Power-Battery Power (p.u.)")
    ax1.set_xlabel("Real Time")
    ax1.xaxis.set_major_locator(mdates.HourLocator(interval=5))
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax1.tick_params(axis='x', rotation=45)
    ax1.grid(True, which='both', linestyle='--', linewidth=2)

    ax2 = plt.subplot(1, 3, 2)
    ax2.plot(t, grid_actions, 'g-', label="Grid Power (p.u.)", linewidth=3)
    ax2.set_ylabel("Grid Power (p.u.)")
    ax2_r = ax2.twinx()
    ax2_r.plot(t, tariffs, 'm-', label="Tariff ($)", linewidth=3)
    ax2_r.set_ylabel("Tariff ($)")
    ax2.set_xlabel("Real Time")
    ax2.xaxis.set_major_locator(mdates.HourLocator(interval=5))
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax2.tick_params(axis='x', rotation=45)
    ax2.grid(True, which='both', linestyle='--', linewidth=2)

    ax3 = plt.subplot(1, 3, 3)
    ax3.plot(t, load_demands, 'y-', label="Load Demand (p.u.)", linewidth=3)
    ax3.set_ylabel("Load Demand (p.u.)")
    ax3_r = ax3.twinx()
    ax3_r.plot(t, load_deferrals, 'c-', label="Load Deferral (p.u.)", linewidth=3)
    ax3_r.set_ylabel("Load Deferral (p.u.)")
    ax3.set_xlabel("Real Time")
    ax3.xaxis.set_major_locator(mdates.HourLocator(interval=5))
    ax3.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax3.tick_params(axis='x', rotation=45)
    ax3.grid(True, which='both', linestyle='--', linewidth=2)

    # Legend with increased font size
    lns = ax1.get_legend_handles_labels()[0] + ax1_r.get_legend_handles_labels()[0] + \
          ax2.get_legend_handles_labels()[0] + ax2_r.get_legend_handles_labels()[0] + \
          ax3.get_legend_handles_labels()[0] + ax3_r.get_legend_handles_labels()[0]
    labs = ax1.get_legend_handles_labels()[1] + ax1_r.get_legend_handles_labels()[1] + \
           ax2.get_legend_handles_labels()[1] + ax2_r.get_legend_handles_labels()[1] + \
           ax3.get_legend_handles_labels()[1] + ax3_r.get_legend_handles_labels()[1]
    ax3.legend(lns, labs, loc='upper center', bbox_to_anchor=(-0.9, -0.25), ncol=6, fontsize=36)

    plt.tight_layout()
    plt.savefig("evaluation.pdf", format="pdf", bbox_inches='tight')
    plt.show()

# Call the function to plot the data
plot_evaluation(df)
```
